```python
import pandas as pd
import numpy as np

df = pd.read_csv("train.csv")
df
```

```python
df.info()
```

```python
def fillNaMode(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])

columns = ['Age','Cabin','Embarked']
fillNaMode(columns)
```

```python
def changeFloattoInt64(cols):
    for i in cols:
        df[i] = df[i].astype('int64')

columns = ['Age','Fare']
changeFloattoInt64(columns)
```

```python
from sklearn.preprocessing import LabelEncoder

def encodeCols(cols):
    for i in cols:
        temp = pd.DataFrame({i:df[i].unique()})
        data_LE = LabelEncoder()
        data_LE.fit(np.ravel(temp))
        
        df[i] = data_LE.transform(df[i])

columns = ['Name','Sex','Ticket','Cabin','Embarked']
encodeCols(columns)
```

```python
df.info()
```

# Training

```python
X = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000, random_state=42)
model.fit(X, y)
```
