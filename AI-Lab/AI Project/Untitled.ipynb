## Dealing with null values

```python
def fillNaMean(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mean())

columns = ['Rating', 'Processor_gen', 'Core_per_processor', 'Total_processor', 'Execution_units', 'Low_Power_Cores', 'Threads', 'Graphics_GB', 'Display_size_inches', 'ppi']
fillNaMean(columns)
```

```python
def fillNaMode(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])

columns = ['Name', 'Brand', 'Processor_brand', 'Processor_brand', 'Processor_variant', 'RAM_type', 'Storage_type', 'Graphics_name', 'Graphics_brand', 'Graphics_integreted', 'Operating_system']
fillNaMode(columns)
```

## Dropping Irrelevant Columns

```python
df.drop('Graphics_integreted', axis=1, inplace=True)
df.drop('Rating', axis=1, inplace=True)
df.drop('Core_per_processor', axis=1, inplace=True)
df.drop('Total_processor', axis=1, inplace=True)
df.drop('Execution_units', axis=1, inplace=True)
df.drop('Low_Power_Cores', axis=1, inplace=True)
df.drop('Energy_Efficient_Units', axis=1, inplace=True)
df.drop('Threads', axis=1, inplace=True)
df.drop('Horizontal_pixel', axis=1, inplace=True)
df.drop('Vertical_pixel', axis=1, inplace=True)
df.drop('ppi', axis=1, inplace=True)
```

```python
df.info()
```

```python

def changeFloattoInt64(cols):
    for i in cols:
        df[i] = df[i].astype('int64')

columns = ['Processor_gen', 'Graphics_GB', 'Display_size_inches','Touch_screen']
changeFloattoInt64(columns)
```

```python
df.info()
```

```python
df_clean = df.copy()
```

```python
def encodeCols(cols):
    for i in cols:
        temp = pd.DataFrame({i:df[i].unique()})
        data_LE = LabelEncoder()
        data_LE.fit(np.ravel(temp))

        df_clean[i] = data_LE.transform(df[i])

columns = ['Name', 'Brand', 'Processor_brand', 'Processor_name', 'Processor_variant', 'RAM_type', 'Storage_type', 'Graphics_name', 'Graphics_brand', 'Operating_system']
encodeCols(columns)
```

# Data Splitting / Train-Test Split

```python
X = df_clean.drop(columns=["Name", "Price"])
Y = df_clean["Price"]

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

print("Training set size:", X_train.shape)
print("Test set size:", X_test.shape)
```

# Training / Applying Classifier / Accuracy:

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print(f"RÂ² Score (Accuracy): {r2:.2f}")
print(f"Mean Absolute Error: {mae:.2f}")
```
