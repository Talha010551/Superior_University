# MiniProject 1:

```python
def fizz_buzz(n):
    for i in range(1, n + 1):
        if i % 3 == 0 and i % 5 == 0:
            print("Fizz Buzz")
        elif i % 3 == 0:
            print("Fizz")
        elif i % 5 == 0:
            print("Buzz")
        else:
            print(i)

fizz_buzz(100)
```

# MiniProject 2:

```python
movies = [
    ("Eternal Sunshine of the Spotless Mind", 20000000),
    ("Memento", 9000000),
    ("Requiem for a Dream", 4500000),
    ("Pirates of the Caribbean: On Stranger Tides", 379000000),
    ("Avengers: Age of Ultron", 365000000),
    ("Avengers: Endgame", 356000000),
    ("Incredibles 2", 200000000)
]

total_movies = int(input("How many movies do you want to add? "))
for _ in range(total_movies):
    name = input("Enter movie name: ")
    budget = int(input("Enter movie budget: "))
    movies.append((name, budget))


average_budget = sum(budget for _, budget in movies) / len(movies)
print(f"\nThe average budget is: ${average_budget:,.2f}\n")


high_budget_movies = [(name, budget) for name, budget in movies if budget > average_budget]
for name, budget in high_budget_movies:
    print(f"{name} had a budget of ${budget:,.2f}, which is ${budget - average_budget:,.2f} above the average.")


print(f"\nNumber of movies with a budget above average: {len(high_budget_movies)}")
```
