###  Implementing a Custom Hash Function and Analyzing Collisions

```python
import random
import string
import matplotlib.pyplot as plt

def custom_hash(key, size):
    return sum(ord(c) for c in key) % size

def builtin_hash(key, size):
    return hash(key) % size

size = 100
keys = [''.join(random.choices(string.ascii_lowercase, k=5)) for _ in range(1000)]

custom_table = [0] * size
builtin_table = [0] * size

for key in keys:
    custom_index = custom_hash(key, size)
    builtin_index = builtin_hash(key, size)
    custom_table[custom_index] += 1
    builtin_table[builtin_index] += 1

plt.subplot(1, 2, 1)
plt.bar(range(size), custom_table)
plt.title("Custom Hash")

plt.subplot(1, 2, 2)
plt.bar(range(size), builtin_table)
plt.title("Built-in Hash")

plt.show()
```
